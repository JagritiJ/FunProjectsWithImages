<h3> Alpha Blending </h3>
Want to overlay a transparent PNG image over another image?

For This the 2 images have to be of the same size
- Put Logo on Shirt/Branded Clothes 
- Put Watermark on the images
- Put Copyright and other signs

```python
import cv2

# read two images
src1 = cv2.imread('buttons.png', cv2.IMREAD_COLOR)
src2 = cv2.imread('beach.png', cv2.IMREAD_COLOR)
same_size_h = 300
same_size_w = 300
src1 = cv2.resize(src1, (same_size_h, same_size_w))
src2 = cv2.resize(src2, (same_size_h, same_size_w))

# add or blend the images
dst = cv2.addWeighted(src1, 0.5, src2, 0.5, 0.0)

# save the output image
cv2.imwrite('image.png', dst)

fig = plt.figure()
fig.add_subplot(1, 3, 1)
plt.imshow(src1)
fig.add_subplot(1, 3, 2)
plt.imshow(src2)
fig.add_subplot(1, 3, 3)
plt.imshow(dst)
```

<h3> Random Mixing of 2 Images </h3>
Take 2 Images, mix the pixels into a 3rd image. Check the new image

```python
import numpy as np
import cv2
```

```python
img1 = cv2.imread('buddha.jpg')
same_size_h = 200
same_size_w = 200
img1 = cv2.resize(img1, (same_size_h, same_size_w))
img2 = cv2.imread('xmastree.jpg')
img2 = cv2.resize(img2, (same_size_h, same_size_w))

new_img = np.zeros([same_size_h,same_size_h,3])

for h in range(0, same_size_h):
    i=0
    for w in range(0, same_size_w):
        if (i/2==0):
            new_img[h, w] = img2[h, w]
        else:
            new_img[h, w] = img2[h,w]
        i+=1

cv2.imshow("image 1", img1)
cv2.imshow("image 2", img2)
cv2.imshow("mixed image", new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
fig.add_subplot(1, 3, 1)
plt.imshow(img1)
fig.add_subplot(1, 3, 2)
plt.imshow(img2)
fig.add_subplot(1, 3, 3)
plt.imshow(new_img)
```
