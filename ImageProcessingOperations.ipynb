<h3>Important to note is that, every pixel is being resprented in BGR format, [244, 200, 10] and reading a portion of image/slicing comes in format of img[height, width, channel]</h3>


```python
#OpenCv reads by default in BGR format, however matplotlib reads in RGB
bgr = cv2.imread("amazingireland.jpg")
cv2.imshow("bgr", bgr)
cv2.moveWindow("bgr", 0,0)
#Other formats for the images, converting BGR to RGB
# rgb = cv2.cvtColor(BGR2RGB)
# hsv = cv2.cvtColor(BGR2HSV)

height, width, channels = bgr.shape
b, g, r = cv2.split(bgr)

#The above split method will showcase the b, g, r channels individually of every pixel
#The split operation is expensive -- need to check time
# Another way
b1 = bgr[:,:, 0]
g1 = bgr[:,:,1]
b1 = bgr[:,:,2]

#Create a blank space for images or skeleton  or a canvas for holding other images

rgb_split = np.empty([height, width*3, 3], 'uint8')

rgb_split[:, 0:width] = cv2.merge([b,b,b])
rgb_split[:, width:width*2] = cv2.merge([g,g,g])
rgb_split[:, width*2:width*3] = cv2.merge([r,r,r])

cv2.imshow('Channels', rgb_split)
cv2.moveWindow('Channels', 0, height)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

<h3> Binary Images </h3>

```python
#Thresholding can be applied on gray scale or binary images
#Binary image - we give a certain threshold value of a pixel, if the pixel value is greater t
# then assign 1 and if it's less then assign 0.
#This doesn't work very weell on landscape images. Use pics with hard boundary and objects
gray_img= cv2.imread('amazingireland.jpg', 0)
height, width = gray_img.shape[0:2]
cv2.imshow("Gray Image", gray_img)

#create a skeleton image of the above shape
binary_img = np.zeros([height, width, 1], 'uint8')

random_thresh = 85
#This method is making the img totally black
for row in range(0, height):
    for col in range (0, width):
        if gray_img[row][col] >random_thresh:
            binary_img[row][col] ==255
            #make it white

cv2.imshow("Binary Image", binary_img)

#This one makes it proper binary
#Or instead of the 2 for loops, use the opencv method called as threshold
ret, thresh_img = cv2.threshold(gray_img, random_thresh, 255, cv2.THRESH_BINARY)

cv2.imshow("thresh_img", thresh_img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

<h3> Area, Perimeter, Centre and Curvature </h3>

```python
original_img = cv2.imread('amazingireland.jpg', 1)
gray_img= cv2.imread('amazingireland.jpg', 0)
height, width = gray_img.shape[0:2]
cv2.imshow("Gray Image", gray_img)
random_thresh =70
max_value=255

adapthresh_img = cv2.adaptiveThreshold(gray_img, random_thresh, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 115, 1)
cv2.imshow("adapthresh_img", adapthresh_img)

contours, heirarchy = cv2.findContours(adapthresh_img, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

img2 = original_img.copy()
index = -1
thickness = 4
color = (255, 0,0)

objects = np.zeros([height, width, 3], 'uint8')

for c in contours:
    cv2.drawContours(objects, [c], -1, color, -1)
    # -1 - all contoours
    # -1 thickness - means fill from inside
    
    area = cv2.contourArea(c)
    perimeter = cv2.arcLength(c, True)
    #True - arc is a closed loop
    # to calculate a centre, you need to calculate the moment
    M = cv2.moments(c)
    print(M)
    if M["m00"] != 0:
        cx = int(M["m10"] / M["m00"])
        cy = int(M["m01"] / M["m00"])
    else:
    # set values as what you need in the situation
        cx, cy = 0, 0
    
    cv2.circle(objects, (cx,cy), 4, (0,0,255), -1)
    
    print("area", area)
    print("perimeter", perimeter)

cv2.imshow("contoured", objects)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Thresholding can sometime combine muktiple objects together, so it is handy to do canny edge first and then threshold

```python
#Treat Fuzzy/Noisy Image
import random

fuzzy_img = cv2.imread('fuzzy.jpg', 1)
height, width, channels = fuzzy_img.shape
fuzzy_gray = cv2.imread('fuzzy.jpg', 0)

blur = cv2.GaussianBlur(fuzzy_gray, (3,3), 0)

thresh = cv2.adaptiveThreshold(blur, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 205, 1)

#Taking the INV to get the foreground images which are non-whoite

cv2.imshow("binary", thresh)

contours, _ = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
print(len(contours))

filtered = []
for c in contours:
    if cv2.contourArea(c)<1000: continue
    filtered.append(c)    

print(len(filtered))

objects = np.zeros([height, width, 3], 'uint8')
for c in filtered:
    col = (random.randint(0,255), random.randint(0,255), random.randint(0,255))
    
    cv2.drawContours(objects, [c], -1, col, -1)
    area = cv2.contourArea(c)
    perimeter = cv2.arcLength(c, True)
    print(area, perimeter)

cv2.imshow("objects", objects)    
cv2.waitKey(0)
cv2.destroyAllWindows()
```

<h3> Template Matching </h3>

Basically a given image is slided over the template image, if there is a perfect match then it is considered white and perfect non-match is black, rest are colors in between.

Doesn't work well:
1. Not scale invariant
2. Not rotation ..
3. requires having well-formed template and making input assumptions

```python
template = cv2.imread('template.jpg', 0)
#this is the particular image we want to match
frame = cv2.imread('friends.jpg', 0)

result = cv2.matchTemplate(frame, template, cv2.TM_CCOEFF_NORMED)

min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)

cv2.imshow("result", result)    
cv2.waitKey(0)
cv2.destroyAllWindows()
```
